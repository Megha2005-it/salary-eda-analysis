```python
import matplotlib.pyplot as plt
```

```python
plt.hist(df["salary"])
plt.xlabel("Salary")
plt.ylabel("NO. of Employees")
plt.title("Salary Distribution")
plt.show()
```

How many employees fall into different salary ranges? 

Observation:
The salary distribution is right-skewed(A few high earners stretch the salary range upward), with most employees earning in the mid-salary range and a few high-salary outliers.
The presence of high earners pulls the mean above the median, which is typical for salary data.

Here the bar shows the count of employees in that range of salary.

```python
df.boxplot(column="salary", by="company")
plt.show()
```

Observation:
Facebook shows the highest median salary among the companies, followed by Google, while ABC Pharma has the lowest median pay. Google exhibits the widest salary spread, indicating high variability in compensation, whereas ABC Pharma salaries are more tightly clustered at the lower end.

```python
df.boxplot(column="salary", by= "job")
plt.show()
```

Observation:
Business Managers have the highest median salary, while Sales Executives have the lowest median salary, with Computer Programmers falling in between. Sales Executives show the highest salary variation, indicating a wide range of pay, whereas Business Managers have the lowest spread, suggesting more consistent compensation.

```python
df.boxplot(column="salary", by="degree")
plt.show()
```

Observation:
Employees with a master’s degree have a higher median salary compared to those with a bachelor’s degree, indicating higher typical pay. However, bachelor’s degree salaries show slightly greater variation, and there is noticeable overlap between the two groups, suggesting that degree alone does not fully determine salary.

Overall Insight

Salary differences in the dataset are influenced by a combination of company, job role, and degree, rather than any single factor. While higher education and certain companies are associated with higher typical salaries, variability within roles and overlap across categories highlight the importance of multiple contributing factors.
